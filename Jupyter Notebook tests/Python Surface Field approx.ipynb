```python
from geomdl import *
import struct
import numpy as np
import scipy
```

```python
file = open("D:/Documents/RLUtilities/assets/soccar/soccar_navigation_nodes.bin",
            "rb")
x = []
y = []
z = []
byte = file.read(4)
while(byte):
    x.append(list(struct.unpack('f', byte)))
    byte = file.read(4)
    y.append(struct.unpack('f', byte)[0])
    byte = file.read(4)
    z.append(struct.unpack('f', byte)[0])
    byte = file.read(4)
```

```python
x[0][0]
```

```python
points = []
for i in range(0, len(x)):
    buff = [x[i][0], y[i][0], z[i][0]]
    points.append(buff)
```

```python
p = np.asarray(points)
```

```python
# Get sorting by x
xind = np.argsort(p[:,0], axis=0)
sortx = np.take_along_axis(p[:,0], xind, axis=0)
new_sorted = []
# Sort points by x first
for i in xind:
    new_sorted.append(p[i])

# get sorting by y
```

```python
ns = sorted(p, key=lambda u:(u[2], u[1], u[0]))
print(ns)
```

```python
sur = geomdl.fitting.approximate_surface(ns, 100, 100, 2, 2)
```

```python
sur.derivatives(0,0, order=1)
```

```python
sur_points = sur.evalpts
```

```python
np.asarray(sur_points) - np.asarray(ns)
```

```python
newfile = open('interpolated_points.bin', 'wb')
for i in sur_points:
    for j in i:
        s = struct.pack('f', j)
        print(j)
        newfile.write(s)
```
