```
nodes = BinaryReadList["D:/Documents/RLUtilities/assets/soccar/soccar_navigation_nodes.bin", 
{"Real32","Real32","Real32"}];
normals = BinaryReadList["D:/Documents/RLUtilities/assets/soccar/soccar_navigation_normals.bin", 
{"Real32","Real32","Real32"}];
Graphics3D[
Table[Line[{nodes[[i]], nodes[[i]] + normals[[i]] * 150}], {i, 1, Length[nodes]}]
]
monomials[n_] := MonomialList[(1+x+y+z)^n] / MonomialList[(1+x+y+z)^n]/.{x\[ShortRightArrow]1, y\[ShortRightArrow]1, z\[ShortRightArrow]1}
```

```
f[{x_, y_, z_}] = Evaluate[monomials[4]];
df[{x_, y_, z_}] = Evaluate[Transpose[Grad[f[{x, y, z}], {x, y, z}]]];
n = Length[nodes];
q = Length[f[{0, 0, 0}]];
F = f /@ (nodes / 1000);
dF = ArrayReshape[df /@ (nodes / 1000), {3 * n, q}];
g1 = ConstantArray[0, Length[F]];
g2 = Flatten[normals];
alpha = 100;
A = Transpose[F].F + alpha * Transpose[dF].dF;
b = alpha * Transpose[dF].g2;
c = Quiet[LinearSolve[A, b]];
Show[
Graphics3D[Point /@ (nodes / 1000)],
ContourPlot3D[c.f[{x,y,z}] == 0, {x, -5, 0}, {y, -6, 6}, {z, -0.25, 2.25}]
]
```
